```python
df['date_added']=pd.to_datetime(df['date_added'])
df.head()
```

```python
df.fillna({'rating':'Unavailable','cast':'Unavailable','country':'Unavailable','director':'Unavailable'},inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df.date_added.isnull()]
```

```python
most_recent_date=df['date_added'].max()
df.fillna({'date_added': 'most_recent_date'},inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df.duration.isnull()]
```

```python
df[df.director=='Louis C.K.']
```

```python
df.loc[df['director']== 'Louis C.K.','duration']=df['rating']
df[df.director=='Louis C.K.'].head()
```

```python
#loc helps us access the specific columns
df.loc[df['director']== 'Louis C.K.','rating']='Unavailable'
df[df.director=='Louis C.K.'].head()
```

# Visualizations

```python
#Visualizations
#Type of shows watched on netflix
df.type.value_counts()
```

```python
sns.countplot(x='type',data=df)
plt.show('Count Vs Type of Shows')
```

```python
df['country'].value_counts().head(10)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='country', order=df['country'].value_counts().index[0:10],data=df)
plt.title('Country wise content on Netflix')
```

```python
movie_countries=df[df['type']=='Movie'] 
shows_countries=df[df['type']=='TV Show']
```

```python
#Top 10 Countries
plt.figure(figsize=(12,6))
sns.countplot(x='country', order=df['country'].value_counts().index[0:10],data=movie_countries)
plt.title('Top 10 Movies Countries producing movies on Netflix')

plt.figure(figsize=(12,6))
sns.countplot(x='country', order=df['country'].value_counts().index[0:10],data=shows_countries)
plt.title('Top 10 Movies Countries producing TV Shows on Netflix')

```

```python
df['rating'].value_counts().head(10)
```

```python
#Rating of shows of Netflix Vs Count Analysis

plt.figure(figsize=(12,6))
sns.countplot(x='rating', order=df['rating'].value_counts().index[0:10],data=df)
plt.title('Rating of shows of Netflix Vs Count')
```

```python
df['release_year'].value_counts()[:20]
```

```python
#Content Release in Years on Netflix Vs Count Analysis
plt.figure(figsize=(12,6))
sns.countplot(x='release_year', order=df['release_year'].value_counts().index[0:20],data=df)
plt.title('Content Release in Years on Netflix Vs Count')
```

```python
#Popular Genre Analysis
plt.figure(figsize=(12,6))
sns.countplot(y='listed_in', order=df['listed_in'].value_counts().index[0:20],data=df)
plt.title('Top 10 Genres on Netflix')
```
